```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import RidgeClassifier
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('final_feat.txt', sep='\t')
df.drop(df.columns[[0]], axis=1, inplace=True)
features = df.as_matrix()
```

```python
np.random.shuffle(features)
train = features[:18000]
test = features[18000:]
```

```python
rf = RandomForestClassifier(n_estimators=50)
rf.fit(train[:,1:], train[:,0])
rf.score(train[:,1:], train[:,0]), rf.score(test[:,1:], test[:,0])
```

```python
ridge_feat = features.copy()
for i in range(6):
    ridge_feat[1+i] = ridge_feat[1+i] / np.max(ridge_feat[1+i])
np.random.shuffle(ridge_feat)
ridge_ftrain = ridge_feat[:18000]
ridge_ftest = ridge_feat[18000:]
```

```python
ridge = RidgeClassifier(alpha=.1, normalize=True, max_iter=10000)
ridge.fit(ridge_ftrain[:,1:], ridge_ftrain[:,0].astype(int))
ridge.score(ridge_ftrain[:,1:], ridge_ftrain[:,0].astype(int)), ridge.score(ridge_ftest[:,1:], ridge_ftest[:,0].astype(int))
```
