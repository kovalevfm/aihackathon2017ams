```python
import re
import sys
import gzip
import time
import nltk
import cPickle
import argparse
import mimetypes
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from itertools import izip

import nltk.tokenize.moses
from nltk.tag.stanford import StanfordNERTagger
# CHUNK_SIZE = 50000
CHUNK_SIZE = 500
NOT_NER_TAG = 'O'

def open_or_gzipopen(fname):
    _, arch = mimetypes.guess_type(fname)
    return gzip.open(fname) if arch == 'gzip' else open(fname)

# german_st_ner = StanfordNERTagger('StanfordNER/classifiers/german.conll.hgc_175m_600.crf.ser.gz', 'StanfordNER/stanford-ner.jar')

# class StanfordNer(object):
#     def __init
def get_stanford_ner(lang='en'):
    if lang == 'de':
        return StanfordNERTagger('/home/fkovalev/dev/mltext/StanfordNER/classifiers/german.conll.hgc_175m_600.crf.ser.gz', '/home/fkovalev/dev/mltext/StanfordNER/stanford-ner.jar')
    if lang ==  'en':
        return StanfordNERTagger('/home/fkovalev/dev/mltext/StanfordNER/classifiers/english.all.3class.distsim.crf.ser.gz', '/home/fkovalev/dev/mltext/StanfordNER/stanford-ner.jar')


class PatchedMosesTokenizer(object):
    def __init__(self, lang='en'):
        self.del_zwsp = re.compile(ur'(?u)[\s\u200b]+')
        self.tok = nltk.tokenize.moses.MosesTokenizer(lang=lang)
        self.detok = nltk.tokenize.moses.MosesDetokenizer(lang=lang)
    def tokenize(self, sentence):
        return [self.detok.unescape_xml(w) for w in self.tok.tokenize(self.del_zwsp.sub(u' ', sentence))]

def find_intervals(tagged):
    current_tp = None
    current_idx, beg_idx = 0, 0
    sentence, intervals = [], []
    for word, tp in tagged:
        if tp != current_tp:
            if current_tp != NOT_NER_TAG and current_idx > 0:
                # store
                intervals.append((beg_idx, current_idx - 1))
            if tp != NOT_NER_TAG:
                # start new interval
                beg_idx = current_idx
            current_tp = tp
        else:
            pass # move
        current_idx += len(word) + 1
        sentence.append(word)
    if current_tp != NOT_NER_TAG and current_idx > 0:
        intervals.append((beg_idx, current_idx - 1))
    return (u' '.join(sentence)), intervals

def get_entities(tagged):
    current_tp = None
    words = []
    for word, tp in tagged:
        if tp != current_tp:
            if current_tp != NOT_NER_TAG and current_tp is not None:
                # store
                yield (u' '.join(words), current_tp)
                words = []
            if tp != NOT_NER_TAG:
                # start new interval
                words.append(word)
            current_tp = tp
        else:
            if tp != NOT_NER_TAG:
                words.append(word)
    if current_tp != NOT_NER_TAG and current_tp is not None:
        yield (u' '.join(words), current_tp)

    
```
